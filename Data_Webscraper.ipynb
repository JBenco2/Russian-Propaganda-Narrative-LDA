### Loading in missing article links

```python
missingLinks = pd.read_excel('Missing March Link.xlsx')
```

```python
missingLinks
```

```python
driver = webdriver.Chrome(executable_path=r"E:\Chromedriver\chromedriver_win32_chrome83\chromedriver.exe")
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}        
articlelist=[]
for item in missingLinks['link']:
    driver.get(item)
    #driver.switch_to.frame('main_frame')
    #driver.Manage().timeouts().implicitWait(3,TimeUnit.SECONDS)
    driver.implicitly_wait(3)
    
    element1 = driver.find_element(By.XPATH,'//*[@id="news"]/div[1]/div/div[1]/div/h1').text
    
    #wait = WebDriverWait(driver, Duration.ofSeconds(10));
    #wait.until(ExpectedConditions.presenceOfElementLocated(element1));
    
    
    element2 = item
    
    
    try:
        if driver.find_element(By.XPATH,'//*[@id="news"]/div[1]/div/div[1]/div/div[1]/div[2]/span/dateformat/span') != None:
            element3 = driver.find_element(By.XPATH,'//*[@id="news"]/div[1]/div/div[1]/div/div[1]/div[2]/span/dateformat/span').text
        elif driver.find_element(By.XPATH,'//*[@id="news"]/div[1]/div/div[1]/div/div[1]/div/span/dateformat/span') != None:
            element3 = driver.find_element(By.XPATH, '//*[@id="news"]/div[1]/div/div[1]/div/div[1]/div/span/dateformat/span').text
        else:
            element3 = ''
    except NoSuchElementException:
        pass
    
    
    try:
        if driver.find_element(By.XPATH,'//*[@id="news"]/div[1]/div/div[2]/div') != None:
            element4 = driver.find_element(By.XPATH,'//*[@id="news"]/div[1]/div/div[2]/div').text
        elif driver.find_element(By.XPATH,'//*[@id="news"]/div[1]/div/div[3]/div') != None:
            element4 = driver.find_element(By.XPATH,'//*[@id="news"]/div[1]/div/div[3]/div').text
        else:
            element4 = ''
    except NoSuchElementException:
        pass
    
    
    article = {
        'title':element1,
        'link':element2,
        'uploadtime':element3,
        'content':element4,
    }
    articlelist.append(article)
    #driver.quit()
```
